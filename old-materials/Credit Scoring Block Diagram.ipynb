```python
dot = pydot.Dot()                                                           
dot.set('rankdir', 'TB')                                                    
dot.set('concentrate', True)                                                
dot.set_node_defaults(shape='record')

# Create a graph and set defaults

node = pydot.Node('E', label="Employment and Assets")
dot.add_node(node)

node = pydot.Node('H', label="Alternative Credit History")
dot.add_node(node)

node = pydot.Node('A', label="Borrower's Application")
dot.add_node(node)

node = pydot.Node('B', label="Data Brokers")
dot.add_node(node)

node = pydot.Node('T', label="Traditional data")
dot.add_node(node)

node = pydot.Node('N', label="Nontraditional data\n(Social)")
dot.add_node(node)

node = pydot.Node('F', label="FCRA")
node.set("color", "blue")
dot.add_node(node)

node = pydot.Node('S', label="Credit Report")
dot.add_node(node)

node = pydot.Node('ECOA', label="ECOA")
node.set("color", "blue")
dot.add_node(node)


node = pydot.Node('D', label="Lender Decision")
dot.add_node(node)

dot.add_edge( pydot.Edge('E','A'))
dot.add_edge( pydot.Edge('H','A'))
dot.add_edge( pydot.Edge('A','D'))

dot.add_edge( pydot.Edge('B','S'))
dot.add_edge( pydot.Edge('F','S'))
dot.add_edge( pydot.Edge('N','S'))
dot.add_edge( pydot.Edge('T','S'))


dot.add_edge( pydot.Edge('B','D'))
dot.add_edge( pydot.Edge('ECOA','D'))
dot.add_edge( pydot.Edge('S','D'))
```

```python
for n in dot.get_nodes():
    #n.set('style', 'filled')
    #n.set('fillcolor', 'aliceblue')
    n.set('fontsize', '10')
    n.set('fontname', 'Trebuchet MS, Tahoma, Verdana, Arial, Helvetica, sans-serif')

SVG( data=dot.create_svg() )
```
